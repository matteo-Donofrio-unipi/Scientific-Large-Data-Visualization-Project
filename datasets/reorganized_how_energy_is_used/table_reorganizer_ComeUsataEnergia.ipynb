```python
#CONVERSION FROM RAW TABLE TO MODIFYED TABLE
import pandas as pd 
for i in range(1,40,1):
    df1 = pd.read_csv("nrg_bal_s_"+str(i)+"_Data.csv")
    name=df1.iloc[0]['GEO']
    print(name)
    df1=df1.drop(['GEO', 'UNIT','Flag and Footnotes', 'SIEC'], axis=1)
    df2 = df1.pivot_table(index='NRG_BAL', columns='TIME', values ='Value', aggfunc='first',fill_value=0)
    df2=df2.replace({':': 0})
    
    #Rimozione del prefisso da ogni stringa nell'index
    listIndex =[]
    for i in range (len(df2.index)):
        listIndex.append( df2.index[i].replace('Final consumption - ',''))
    
    df2['Use']=listIndex
    
    df2 =df2.set_index('Use')
    
    df2.to_csv('reorganized/'+name+'_how_energy_is_used.csv')
#df1= pd.read_csv('reorganized/Italy_how_energy_is_used.csv', index_col=0)
```
