```python
import pandas as pd
for i in range(1,40,1):
    df1 = pd.read_csv("nrg_bal_peh_"+str(i)+"_Data.csv")
    name=df1.iloc[0]['GEO']
    df1=df1.drop(['GEO', 'UNIT','Flag and Footnotes', 'NRG_BAL'], axis=1)
    df2 = df1.pivot_table(index='SIEC', columns='TIME', values ='Value', aggfunc='first',fill_value=0)
    df2=df2.replace({':': 0})
    df2.to_csv('reorganized/'+name+'_how_produce_energy.csv')
```
